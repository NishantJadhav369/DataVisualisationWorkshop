**Import numpy and pandas.
Create df from dictionary**

```python
import numpy as np
import pandas as pd
```

```python
# Create dataframe
data = {'Company':['GOOGLE','GOOGLE','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
df
```

**GROUPBY - mean, max, describe, transpose, unique, value_counts , apply**

```python
dfgrp =df.groupby('Company')
```

```python
dfgrp.mean()
```

```python
dfgrp.max()
```

```python
dfgrp.describe()
```

```python
dfgrp.describe().transpose()
```

```python
dfgrp.describe().transpose()['GOOGLE']
```
